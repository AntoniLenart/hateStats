```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
try:
    path = '../data/complete_user_data.csv'
    df = pd.read_csv(path, index_col=0)
except FileNotFoundError as fnfError:
    print(fnfError)
except:
    print(f'Unknown error has occured while loading DataFrame from {path}')
```

```python
WN8_COLORS = ['#000000', '#cd3333', '#d77900', '#d7b600', '#6d9521', '#4c762e', '#4a92b7', '#83579d', '#5a3175']
WN8_BINS = [0, 300, 600, 900, 1250, 1600, 1900, 2350, 2900, 4000]
WN8_NAMES = ['Very Bad', 'Bad', 'Under Average', 'Average', 'Good', 'Very Good', 'Great', 'Unique', 'Super Unique']
```

```python
df['wn8'] = df['wn8'].str.replace(',', '.')
df['wn8'] = df['wn8'].astype(float)
df['wn8']
```

### Data visualizations

```python
enum = enumerate(WN8_COLORS)
for i, a in enum:
    print(a)
```

```python
# Histogram of wn8 scores
fig, ax = plt.subplots(figsize=(10,6))
data = df['wn8']

N, bins, patches = ax.hist(data)
for i, color in enumerate(WN8_COLORS):
    patches[i].set_facecolor(color)

ax.set(title='WN8 Histogram',
       xlabel='WN8 Score',
       ylabel='Number of players')

plt.savefig('../data/figures/wn8_hist.png')
```

```python
# Average xp per battle for different wn8 scores
groups = df.groupby(pd.cut(df['wn8'], WN8_BINS), observed=True)
data = groups['battle_avg_xp'].median()
data.index = WN8_NAMES
data.plot(kind='barh',
          title='Medians of players xp per battle for different skill levels',
          xlabel='Experience',
          ylabel='Player skill based on WN8',
          color=WN8_COLORS,
          figsize=(10,6))
plt.savefig('../data/figures/xp_per_skill.png')
```

```python
# Battles played histogram
df['battles'].plot(kind='hist',
                   title='Histogram of battles played',
                   xlabel='Number of battles',
                   ylabel='Number of players',
                   xlim=[0,50000],
                   bins=50,
                   figsize=(10,6),
                   color='lightblue')
plt.savefig('../data/figures/battle_hist.png')
```

```python
# Average battle survival pie chart
fallen_battles = df['battles']-df['survived_battles']
table = pd.DataFrame(data=[fallen_battles, df['survived_battles']])
table = table.T
table.columns = ['Fallen battles', 'Survived battles']
table.mean().plot(kind='pie',
                  title='Averaged battle survival rate',
                  colors=['darkgrey', 'lightgreen'],
                  autopct='%.2f%%',
                  figsize=(10,6))
plt.savefig('../data/figures/battle_survival_rate.png')
```
